```python
#read a file that is tab seperated
products = pd.read_csv('products-2.csv', sep='\t')
```

```python
#inspect our data
products.head() #the first 5 rows
```

```python
#summary stats
#basic statistical summary of all numeric values in the data set
products.describe()
```

```python
#dispalying subsets of our dataset
products['ListPrice'] #this returns a pandas series objects(this is not a data frame)
```

```python
products[['ListPrice']] #this returns a dataframe (double square brackets)
```

```python
products[['ProductID','ListPrice','Color','Weight']] #multiple columns
```

```python
#list the columns in the data frame
products.columns
```

```python
#subsets of our data based on certain condition
# show all the products that are black(productID, color, weight)
products[products['Color']=='Black']
```

```python
#subsets of our data based on certain condition
# show all the products that are black(productID, color, weight)
products[products['Color']=='Black'][['ProductID','Color','Weight']]
```

```python
#subsets of our data based on certain condition
# show all the products that are black(productID, color, weight)
# another way of doing it
black_products=products[products['Color']=='Black']
```

```python
black_products[['ProductID','Color','Weight']]
```

```python
#list the productID, name, and reorderpoint of all products with a list price greater than 500
products[products['ListPrice']>500][['ProductID','Name','ReorderPoint']]
```

```python
#check how many values there are of each type
products['Color'].value_counts()
```

```python
#how to create a new column
products['DiscountPrice']=products['ListPrice']*0.5
```

```python
products.head()
```

```python
#remove a column
products.drop(columns=['DiscountPrice'], inplace=True)
```
